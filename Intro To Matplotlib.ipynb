### Two options for subplots






```python
#OPTION 1
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows = 2,
                                           ncols = 2,
                                           figsize = (10,5))

#Plot to different axes each
ax1.plot(x, x/2)
ax2.scatter(x, np.sin(x))
ax3.bar(sales_products.keys(), sales_products.values())
ax4.plot(x,x**2)
```

```python
#OPTION 2
fig, ax = plt.subplots(nrows = 2,
                       ncols = 2,
                       figsize = (10,5))

#PLot to each different index
ax[0,0].plot(x, x/2);
ax[0,1].scatter(x, np.sin(x));
ax[1,0].bar(sales_products.keys(), sales_products.values());
ax[1,1].plot(x, x**2);
```

__Which one should you use? (pyplot vs matplotlib OO method)__

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
#pyplot method
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target');

```

```python
#OO method with mixed pyplot
%matplotlib inline
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10,6))

over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);

#ax.set_xlim(45,100)
```

```python
#OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

#creating plot
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);
#customizing plot
ax.set(title="Heart Disease and cholestrol levels",
       xlabel="Age",
       ylabel="Choleastrol");

#creating a legend
ax.legend(*scatter.legend_elements(), title="Target")

#creating a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle='--')
```

```python
#creating subplots
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10));
#creating ax0 plot
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);
#customizing ax0 plot
ax0.set(title="Heart Disease and Cholestrol levels",
       xlabel="Age",
       ylabel="Choleastrol");

#creating a legend in ax0
ax0.legend(*scatter.legend_elements(), title="Target");

#creating a horizontal line in ax0
ax0.axhline(over_50["chol"].mean(),
           linestyle='--');



#creating ax1 plot
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"]);
#customizing ax0 plot
ax1.set(title="Heart Disease and Cholestrol levels",
       xlabel="Age",
       ylabel="Max Heart Rate");

#creating a legend in ax0
ax1.legend(*scatter.legend_elements(), title="Target");

#creating a horizontal line in ax0
ax1.axhline(over_50["thalach"].mean(),
           linestyle='--');


#add title to whole figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
